## Classification of English Words by IPA Pronunciation

```python
# Import packages
import pandas as pd
import more_itertools as mit
import oxford as ox
import pprint
from oxford import Word
from nltk.stem.wordnet import WordNetLemmatizer
from pattern.en import lemma
import time
```

```python
# Data cleaning and Data Preparation
df_text = pd.read_csv("C://Users//acer//OneDrive//DataDisca//21_10_2021 Task 02//words.txt",
                      header = None,
                      sep='\n'
                     )
#print(df_text)
```

```python
for i in df_text[0]:
    line = i
    #print(line)
    test = line.lower().split()
    test2 = sorted([j for j in test if j.isalpha() ])
    #print(test2)
```

```python
df_w = set(mit.collapse(test2))
#df_w
```

```python
df_w1 = []
for i in df_w:
       df_w1.append(lemma(i))
```

```python
#create Word list
df_w = sorted(df_w1)
df_w = set(mit.collapse(df_w))
df_w = sorted(list(df_w))
```

```python
#create Word data frame
df = pd.DataFrame(df_w, columns=["Word"])
```

```python
#save Word data frame as a CSV file
df.to_csv('C://Users//acer//Desktop//oxford-dictionary-api//df.csv')  
```
